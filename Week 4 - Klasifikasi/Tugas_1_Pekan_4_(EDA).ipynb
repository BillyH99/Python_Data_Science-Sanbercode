```python
# code here
df = df[['DewPointAvgF','HumidityAvgPercent','WindAvgMPH', 'TempAvgF']].astype('float')
```

```python
df = df.sample(1000)
x = df['HumidityAvgPercent']
y = df['DewPointAvgF']

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x,y,s=df['WindAvgMPH']*20,c=df['TempAvgF'],cmap='coolwarm')
ax.set_xlabel('Humidity Avegare (%)')
ax.set_ylabel('Dew Point Average(%)')
ax.set_title('Austin Weather')

```

```python
# Fill null value
df = pd.read_csv('austin_weather.csv')
df.head()
df.replace('na',np.nan,inplace=True)
df.replace('-',np.nan,inplace=True)
df=df[['DewPointAvgF','HumidityAvgPercent','WindAvgMPH']].fillna(method='ffill',inplace=True)
x = df['HumidityAvgPercent']
y = df['DewPointAvgF']

fig2, ax2 = plt.subplots(figsize=(12, 6))
ax2.scatter(x,y,s=df['WindAvgMPH']*20,c=df['TempAvgF'],cmap='coolwarm')
ax2.set_xlabel('Humidity Avegare (%)')
ax2.set_ylabel('Dew Point Average(%)')
ax2.set_title('Austin Weather')
```

Dengan menggunakan Interquartile range, kita bisa temukan nilai data outliers dan meremove data outlier tersebut. 

*   Tentukan nilai batas atas dan batas bawah dari Interquartile range tersebut

```python
#Code here
df2['House_Price'] = sorted(df2['House_Price'])
q1 = np.percentile(df2['House_Price'], 25, interpolation = 'midpoint')
q3 = np.percentile(df2['House_Price'], 75, interpolation = 'midpoint')
# print(q1, q3)
IQR = q3-q1
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print('Lb =',lwr_bound,', Ub =',upr_bound)

# Upper bound
upper = np.where(df2['House_Price'] >= upr_bound)

# Below Lower bound
lower = np.where(df2['House_Price'] <= lwr_bound)

df2['House_Price'] = np.where(df2['House_Price']>upr_bound, upr_bound, df2['House_Price'])
#df2['House_Price'].drop(upper[0], inplace = True)
#df2['House_Price'].drop((lower[0]), inplace = True)
plt.boxplot(df2['House_Price'], vert=False)
```
